```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Create a directed graph
G = nx.DiGraph()

# 2. Add nodes (the letters in your graph)
G.add_nodes_from(['u', 'v', 'w', 'x', 'y', 'z'])

# 3. Add edges (the connections between the nodes)
edges = [
    ('u', 'v'),
    ('u', 'x'),
    ('v', 'y'),
    ('y', 'x'),
    ('x', 'v'),
    ('w', 'y'),
    ('w', 'z'),
    ('z', 'z')
]
G.add_edges_from(edges)

pos = nx.shell_layout(G)

# 5. Visualize the graph
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1000, font_size=12, arrows=True)
plt.title("Graph Visualization")
plt.show()
```

```python
vertices = ['u', 'v', 'w', 'x', 'y', 'z']
graph = {
    'u' : ['v', 'x'],
    'v' : ['y'],
    'x' : ['v'],
    'y' : ['x'],
    'w' : ['y', 'z'],
    'z' : ['z']
}
color = {}
parent = {}
```

```python
def color_graph(vertices):
    for vertex in vertices:
        color[vertex] = "white"
        parent[vertex] = None
    return vertices
```

```python
def dfs_visit(graph, vertex):
    color[vertex] = "gray"
    for v in graph[vertex]:
        if color[v] == "white":
            parent[v] = vertex
            dfs_visit(graph, v)
    color[vertex] = "black"
```

```python
def dfs(graph, vertices):
    vertices = color_graph(vertices)
    for vertex in vertices:
        if color[vertex] == "white":
            dfs_visit(graph, vertex)
```
